```python
from pred import Predictor
from pred import sequence_vector
from pred import chemical_vector
from pred import DataDict
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline



y = Predictor()
y.load_data(file="Data/Training/clean_s.csv")
y.process_data(vector_function="sequence", amino_acid="S", imbalance_function="pass", random_data=0)
print(len(y.features[0]))
print(len(y.labels))
```

```python
feat = pd.DataFrame(np.array(y.features).reshape(len(y.features), len(y.features[0])), 
                  columns = ["G", "A", "L", "M", "F", "W", "K", "Q", "E", "S", "P", "V", "I", "C", "Y", "H", "R", "N", "D", "T", "X", "GRAVY", "Instability Index", "Aromatacity"])
labels = pd.DataFrame(np.array(y.labels).reshape(len(y.labels), 1), columns = ["labels"])
print("Labels", labels.shape)
print("Features", feat.shape)
data = pd.concat([feat,labels], axis=1)
print('Data', data.shape)
```
